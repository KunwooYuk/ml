```python
import pandas as pd

file_path = 'cancer.xlsx' 
sheet_all = 'Data'

# pd.read_excel 함수를 사용해 특정 시트를 읽어온다.
df = pd.read_excel(file_path, sheet_name=sheet_all)

df.columns
```

```python
df_selec = df[(df['Mixture']=='Chemical') & (df['Data_Type']=='In Vivo')& (df['Species']=='Mouse')]
df_selec
#Mixture 컬럼이 Chemical, Data_Type 컬럼은 In Vivo인 경우 선택.
```
